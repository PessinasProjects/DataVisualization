```python
import matplotlib.pyplot as plt

```

```python
#first diagram
squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(squares)

plt.show()
```

```python
#comments
squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(squares, linewidth=3)

ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

#set size of teilstrichbeschriftungen
ax.tick_params(axis="both", labelsize=14)

```

```python
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10)

ax.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10)

ax.axis([0, 1100, 0, 1_100_000])
ax.ticklabel_format(style="plain")

plt.show()
```

```python
#einige Farben festlegen
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, color="red", s=10)

ax.axis([0, 1100, 0, 1_100_000])
ax.ticklabel_format(style="plain")

plt.show()
```

```python
#einige Farben festlegen
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, color=(0, .8, 0), s=10)

ax.axis([0, 1100, 0, 1_100_000])
ax.ticklabel_format(style="plain")

plt.show()
```

```python
#colourmap
#einige Farben festlegen
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use("seaborn-v0_8-whitegrid")
fig, ax = plt.subplots()

ax.scatter(x_values, y_values, color="blue", s=10)

ax.axis([0, 1100, 0, 1_100_000])
ax.ticklabel_format(style="plain")

plt.show()
```

```python
#diagrams save automatically
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use("seaborn-v0_8-whitegrid")
fig, ax = plt.subplots()

ax.scatter(x_values, y_values, color="blue", s=10)

ax.axis([0, 1100, 0, 1_100_000])
ax.ticklabel_format(style="plain")


plt.savefig("squares_plot.png", bbox_inches="tight") #instead of plt.show()
```

```python
#Kubikzahlen
#einige Farben festlegen
x_values = range(1, 6)
y_values = [x**3 for x in x_values]

plt.style.use("seaborn-v0_8-whitegrid")
fig, ax = plt.subplots()

ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Reds, s=100)

ax.axis([0, 5, 0, 125])
ax.ticklabel_format(style="plain")

plt.show()
```
