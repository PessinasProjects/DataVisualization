```python

# plotly: Datenvisualisierung, die in einem Browser angezeigt werden soll- automatische Bildschirmanpassung
import plotly.express as px
import plotly.io as pio
from dado import Dado

pio.renderers.default = "browser"

die = Dado()

results = []

for roll_num in range(1000):
    result = die.roll()
    results.append(result)
    
#print(results) alles funktioniert

frequencies = []

poss_results = range(1, die.num_sides + 1)

for value in poss_results:
    frequency = results.count(value)
    
    frequencies.append(frequency)
    
print(frequencies) #wie oft die einzelne Nummer vorkommen
```

```python
die = Dado()

results = []

for roll_num in range(1000):
    result = die.roll()
    results.append(result)

frequencies = []

poss_results = range(1, die.num_sides + 1)

for value in poss_results:
    frequency = results.count(value)
    
    frequencies.append(frequency)
    
title = "Results of rolling one D6 1000 Times" #titel hinzugefüft

labels = { "x": "Results", "y": "Frequency of results"} #lables hinzugefügt

fig = px.bar(x=poss_results, y=frequencies, title=title, labels=labels)

fig.show()
```

```python
#Ergebnisse bei 2 Würfel

die_1 = Dado()
die_2 = Dado()

results = []

for roll_num in range(1000):
    result = die_1.roll() + die_2.roll()
    results.append(result)

frequencies = []

max_results = die_1.num_sides + die_2.num_sides
poss_results= range(2, max_results + 1)

for value in poss_results:
    frequency = results.count(value)
    
    frequencies.append(frequency)
    
title = "Results of rolling two D6 1000 Times"

labels = { "x": "Results", "y": "Frequency of results"} 

fig = px.bar(x=poss_results, y=frequencies, title=title, labels=labels)

fig.update_layout(xaxis_dtick=1) #Alle x Balken werden beschriftet, wirkt auf das fig Object

fig.show()
```

```python
#Ergebnisse bei 2 Würfel uterschiedlicher Flächenzahl

die_1 = Dado()
die_2 = Dado(10)

results = []

for roll_num in range(50000):
    result = die_1.roll() + die_2.roll()
    results.append(result)

frequencies = []

max_results = die_1.num_sides + die_2.num_sides
poss_results= range(2, max_results + 1)

for value in poss_results:
    frequency = results.count(value)
    
    frequencies.append(frequency)
    
title = "Results of rolling one D6 and one d10 50000 Times"

labels = { "x": "Results", "y": "Frequency of results"} 

fig = px.line(x=poss_results, y=frequencies, title=title, labels=labels)

fig.update_layout(xaxis_dtick=1) #Alle x Balken werden beschriftet, wirkt auf das fig Object

fig.show()
```
