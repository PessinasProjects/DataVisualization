```python
# Excercise 16-8 
#Excercise 16-6
#Darstellung der Stärke
from pathlib import Path
import json
try:
    # For Python 3.0 and later
    from urllib.request import urlopen
except ImportError:
    # Fall back to Python 2's urllib2
    from urllib2 import urlopen

from pprint import pprint

import plotly.express as px
import plotly.io as pio


def get_jsonparsed_data(url):
    """
    Receive the content of ``url``, parse it as JSON and return the object.

    Parameters
    ----------
    url : str

    Returns
    -------
    dict
    """
    response = urlopen(url)
    data = response.read().decode("utf-8")
    return json.loads(data)
#
# url is defined above.
#

url = ("https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.geojson")
all_eq_data = get_jsonparsed_data(url)


# Alle Erdbeben im Datensatz analysieren.
all_eq_dicts = all_eq_data["features"]
print(len(all_eq_dicts))

# Take magnitude and place data.
mags, lons, lats, eq_titles = [] ,[], [], []
for eq_dict in all_eq_dicts:
    mags.append(eq_dict["properties"]["mag"])
    lons.append(eq_dict["geometry"]["coordinates"][0])
    lats.append(eq_dict["geometry"]["coordinates"][1])
    eq_titles.append(eq_dict["properties"]["title"])

min_mag = min(mags)
max_mag = max(mags)

#to open the diagram in browser
#Credits: Dr. Pardeep and Dr. Vikas.
pio.renderers.default = "browser"

# Nimm den Titel from json
title = all_eq_data["metadata"]["title"]
#fig = px.scatter_geo(...color_continuous_scale="burg", "blugrn" usw
#print(px.colors.named_colorscales())
#fig.update_layout(coloraxis=dict(cmin=min_mag, cmax=max_mag))
fig = px.scatter_geo(lat=lats, 
                     lon=lons, 
                     size=list(map(lambda x:x-min_mag, mags)), 
                     title=title, 
                     color=list(map(lambda x:x-min_mag, mags)), 
                     color_continuous_scale="Viridis", 
                     labels={"mags":"Magnitude"}, 
                     projection="natural earth",
                     hover_name=eq_titles)
fig.update_layout(coloraxis=dict(cmin=min_mag, cmax=max_mag))
fig.show()

```
