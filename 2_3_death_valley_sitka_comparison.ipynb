```python
from pathlib import Path
import csv
import matplotlib.pyplot as plt
from datetime import datetime

path2 = Path('weather_data/sitka_weather_2021_simple.csv')
lines2 = path2.read_text().splitlines()   
reader2 = csv.reader(lines2) 
header_row2 = next(reader2) 

dates2, highs2, lows2 = [], [], []
for row2 in reader2:
    current_date2 = datetime.strptime(row2[2], '%Y-%m-%d')
    high2 = int(row2[4])
    low2= int(row2[5])
    dates2.append(current_date2)
    highs2.append(high2)
    lows2.append(low2)
    
# print(plt.style.available)
      
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.plot(dates2, highs2, color='violet')
ax.plot(dates2, lows2, color='blue')
ax.set_ylim(0, 140)

path = Path('weather_data/death_valley_2021_simple.csv')
lines = path.read_text().splitlines()   
reader = csv.reader(lines) 
header_row = next(reader)

plt.title('Sitka, 2021', fontsize=20)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
    
dates, highs, lows = [], [], []

for row in reader:
    current_date = datetime.strptime(row[2], '%Y-%m-%d')
    try:
        high = int(row[3])
        low = int(row[4])
    except ValueError:
        print(f"Missing data for {current_date}")
    else:
        dates.append(current_date)
        highs.append(high)
        lows.append(low)
    
    
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.plot(dates, highs, color='red')
ax.plot(dates, lows, color='orange')
ax.set_ylim(0, 140)


plt.title('Death Valley, 2021', fontsize=20)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```
