```python
from pathlib import Path
import csv

import matplotlib.pyplot as plt

from datetime import datetime

first_date = datetime.strptime("2021-07-01", "%Y-%m-%d")
print(first_date)

#initialize the path and split text in lines
path = Path("../weather_data/sitka_weather_07-2021_simple.csv")
lines = path.read_text().splitlines()

#initialize csv.reader, read the header
reader = csv.reader(lines)
header_row = next(reader)

# return column names and its position
for index, column_header in enumerate(header_row):
    print(index, column_header)

#read csv data
#the highest temperatures
dates, highs = [], []
for row in reader:
    current_date = datetime.strptime(row[2], "%Y-%m-%d")
    high = int(row[4])
    dates.append(current_date)
    highs.append(high)

#Stellt diw hchsttemperaturen graphosch da
plt.style.use("seaborn-v0_8")
fig, ax = plt.subplots()
ax.plot(dates, highs, color="red")

# Formatiert die diagramm.
plt.title("Daily High temperatures, July 2021", fontsize=24)
plt.xlabel("", fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Temperature(F)", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)

plt.show()
```

```python
#Niederschlag in Sitja und Death_Valley
from pathlib import Path
import csv

import matplotlib.pyplot as plt

from datetime import datetime

#initialize the path and split text in lines
path = Path("../weather_data/death_valley_2021_full.csv")
lines = path.read_text().splitlines()

#initialize csv.reader, read the header
reader = csv.reader(lines)
header_row = next(reader)

# return column names and its position
for index, column_header in enumerate(header_row):
    print(index, column_header)

#read csv data
#the date, highest abd lowest temperatures
dates, rain_data_dv = [], []
for row in reader:
    current_date = datetime.strptime(row[2], "%Y-%m-%d")
    try:
        rain = float(row[3])   
    except ValueError as e:
        print(f"Missing data for {current_date}")
    else:
        dates.append(current_date)
        rain_data_dv.append(rain)




#initialize the path and split text in lines
path = Path("../weather_data/sitka_weather_2021_full.csv")
lines = path.read_text().splitlines()

#initialize csv.reader, read the header
reader = csv.reader(lines)
header_row = next(reader)

# return column names and its position
for index, column_header in enumerate(header_row):
    print(index, column_header)

#read csv data
#the date, highest abd lowest temperatures
dates_sw, rain_data_sw = [], []
for row in reader:
    current_date = datetime.strptime(row[2], "%Y-%m-%d")
    if current_date in dates: #if current date is in the first csv file
        try:
            rain = float(row[5])   
        except ValueError as e:
            print(f"Missing data for {current_date}")
            # Remove that date from  dates and rain_data_dv lists
            index_ = rain_data_dv.index(current_date)
            rain_data_dv.pop(index_)
            dates.pop(index_)
        else:
            rain_data_sw.append(rain)
            dates_sw.append(current_date)

#check if for all dates exists all information
for index_ in range(len(dates)):
    if dates[index_] not in dates_sw:
        rain_data_dv.pop(index_)
        dates.pop(index_)

#Stellt diw hchsttemperaturen und tiefsttemperaturen graphisch da
plt.style.use("seaborn-v0_8")
fig, ax = plt.subplots()
ax.plot(dates, rain_data_dv, color="red", alpha=.5)
ax.plot(dates, rain_data_sw, color="blue", alpha=.5)
ax.legend(['Death Valley', 'Sitka'])
plt.fill_between(dates, rain_data_dv, rain_data_sw, facecolor="blue", alpha=.1)

# Formatiert die diagramm.
plt.title("Daily Differencies in Rain - 2021 \nDeath Valley vs Sitkey", fontsize=24)
plt.xlabel("", fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Rain()", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)

plt.show()
```
